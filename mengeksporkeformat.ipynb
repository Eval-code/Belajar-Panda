```python
import pandas as pd
df = pd.read_csv('pokemon_data.csv')
```

```python
df.head(5)
```

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
```

```python
df.head(3)
```

```python
col = df.pop('Total')
df.insert(5, 'Total', col)
```

```python
df.head(3)
```

```python
#ini mengubah data csv ke excel
df.to_excel('modifikasi.xlsx', index=False)
```

```python
#merubah lagi dari excel ke csv
df.to_csv('modifikasi.csv', index=False)
```

```python
#merubah dari csv ke txt
df.to_csv('modifikasi.txt', index=False, sep='\t')
```
