```python
import pandas as pd
df = pd.read_csv('pokemon_data.csv')
```

```python
df.head(1)
```

```python
#filter data sesuai user dengan logika and
new_df = df.loc[(df['Type 2'] == "Poison") & (df['Defense'] > 50)]
```

```python
new_df.head()
```

```python
newa_df = df.loc[(df['Type 1'] == "Water" )]
newa_df
```

```python
#filter data harus ada kaitan dengan kata contoh dibawah
new_df = df.loc[df['Name'].str.contains('Mega')]
new_df
```

```python
#filter kata yang harus tidak ada kaitan nya dengan kata, dengan ditambahkan simbol ~
newa_df =df.loc[~df['Name'].str.contains('Mega')]
newa_df
```

```python
#filter menggunakan reguler expresion(regex), dengan mengimport dulu modul re
import re
newwa_df = df.loc[df['Type 1'].str.contains('Fire | Grass', regex=True)]
df.head(10)
```
