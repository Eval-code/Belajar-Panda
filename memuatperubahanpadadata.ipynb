```python
import pandas as pd
df = pd.read_csv('pokemon_data.csv')
```

```python
#menambahkan kolom total
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.head(3)
```

```python
#menghapus kolom total
df.drop(columns=['Total'])
```

```python
#menambahkan kolom total lebih ringkas, df.iloc[:, 4:10] ini merupakan index pengambilan nilai,.sum ini menjumlahkan nilai yg diambil, (axis=1) ini menjumlahkan dengan horizontal jika dia 0 maka vertikal
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
df.head(5)
```

```python
#memindahkan kolom 'Total' ke posisi ke-4 (setelah kolom ke-3)
col = df.pop('Total')     # Hapus kolom 'Total' dan simpan isinya
df.insert(5, 'Total', col)  # Sisipkan kembali di posisi ke-4
df.head(5)
```
